```python
# Task 03
import math
C= 50
H=30
# Q = Square root of [(2 * C * D)/H].

D = input("Enter comma values: ").split(',')
D = [int(x) for x in D]

result = []
for d in D:
    Q = math.sqrt(2*C*d/H)
    result.append(int(Q))

string = ','.join(str(x) for x in result)
print(string)
```

```python
# Task 04
input_str = input("Enter X and Y: ")
X, Y = map(int, input_str.split(','))

# initialising an array filled with zeros because cant apply operations on empty array
# also making it 2D
array = [[0 for j in range(Y)] for i in range(X)]

for i in range(X):
    for j in range(Y):
        array[i][j] = i * j

print(array)
```

```python
# Task 07
input_sentence = input("Enter a sentence: ")

letter_count = 0
digit_count = 0

for char in input_sentence:
    if char.isalpha():
        letter_count += 1
    elif char.isdigit():
        digit_count += 1

result_dict = {
    "LETTERS": letter_count,
    "DIGITS": digit_count
}

print(result_dict)
```

```python
# Task 08
input_numbers = input("Enter comma-separated numbers: ").split(',')

squared_odds = [str(int(num)**2) for num in input_numbers if int(num) % 2 != 0]

result = ','.join(squared_odds)

print(result)
```

```python
# Task 09
balance = 0

transaction_log = input("Enter the transaction log: ")

transactions = transaction_log.split(" ")

for i in range(0, len(transactions), 2):
    if transactions[i] == 'D':
        balance += int(transactions[i + 1])
    elif transactions[i] == 'W':
        balance -= int(transactions[i + 1])

print(balance)
```

```python
# Task 11

num = int(input("Enter an integer: "))

reversed_num = 0
is_negative = False

if num < 0:
    is_negative = True
    num = -num

while num > 0:
    digit = num % 10
    reversed_num = (reversed_num * 10) + digit
    num = num // 10

if is_negative:
    reversed_num = -reversed_num

print("Reversed integer:", reversed_num)
```

```python
# Task 12
import itertools

numbers = [1, 2, 3, 4, 5, 6]

print("Permutations:")
for perm in itertools.permutations(numbers):
    print(perm)

print("\nCombinations:")
for r in range(1, len(numbers) + 1):
    for comb in itertools.combinations(numbers, r):
        print(comb)
```

```python
# Task 01
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('./image_processing_files/house.jpg')

# Apply Gaussian blur 
blurred_image = image.copy()
blurred_image[100:400, 200:500] = cv2.GaussianBlur(blurred_image[100:400, 200:500], (13, 13), 0)

# Convert images to grayscale for edge detection
gray_original = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray_blurred = cv2.cvtColor(blurred_image, cv2.COLOR_BGR2GRAY)

# Apply Canny edge detection on the original and blurred images
edges_original = cv2.Canny(gray_original, 100, 200)
edges_blurred = cv2.Canny(gray_blurred, 100, 200)

# Plot the original image, blurred image, and edge detections
plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB))
plt.title('Blurred Image')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(edges_original, cmap='gray')
plt.title('Edges (Original Image)')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(edges_blurred, cmap='gray')
plt.title('Edges (Blurred Image)')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Task 02
import cv2
import matplotlib.pyplot as plt

# Load square1.jpg and square2.jpg images
square1 = cv2.imread('./image_processing_files/square1.jpg', cv2.IMREAD_GRAYSCALE)
square2 = cv2.imread('./image_processing_files/square2.jpg', cv2.IMREAD_GRAYSCALE)

# Threshold the images to create binary images
_, binary_square1 = cv2.threshold(square1, 127, 255, cv2.THRESH_BINARY)
_, binary_square2 = cv2.threshold(square2, 127, 255, cv2.THRESH_BINARY)

# Perform bitwise OR operation to create an octagon
octagon = cv2.bitwise_or(binary_square1, binary_square2)

# Perform bitwise AND operation to create an 8-pointed star
eight_pointed_star = cv2.bitwise_and(binary_square1, binary_square2)

# Plot the original squares, octagon, and 8-pointed star
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(binary_square1, cmap='gray')
plt.title('Square 1')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(binary_square2, cmap='gray')
plt.title('Square 2')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(octagon, cmap='gray')
plt.title('Octagon')
plt.axis('off')

plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 6))
plt.imshow(eight_pointed_star, cmap='gray')
plt.title('8-Pointed Star')
plt.axis('off')
plt.show()
```

```python
# Task 03
import cv2
import matplotlib.pyplot as plt

# Load the shapes.bmp image
image = cv2.imread('./image_processing_files/shapes.bmp')

# Convert the image to HSV color space
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Define HSV threshold values for each shape (you need to determine these manually)
# For example, below values are placeholders, you need to adjust them based on your image
hsv_min_red = np.array([0, 100, 100])
hsv_max_red = np.array([10, 255, 255])

hsv_min_blue = np.array([90, 100, 100])
hsv_max_blue = np.array([120, 255, 255])

hsv_min_yellow = np.array([20, 100, 100])
hsv_max_yellow = np.array([40, 255, 255])

hsv_min_green = np.array([40, 100, 100])
hsv_max_green = np.array([80, 255, 255])

# Use cv2.inRange() to isolate each shape based on the threshold values
mask_red = cv2.inRange(hsv, hsv_min_red, hsv_max_red)
mask_blue = cv2.inRange(hsv, hsv_min_blue, hsv_max_blue)
mask_yellow = cv2.inRange(hsv, hsv_min_yellow, hsv_max_yellow)
mask_green = cv2.inRange(hsv, hsv_min_green, hsv_max_green)

# Plot the original image and the masks (isolated shapes)
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(mask_red, cmap='gray')
plt.title('Red Shape (Isolated)')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(mask_blue, cmap='gray')
plt.title('Blue Shape (Isolated)')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(mask_yellow, cmap='gray')
plt.title('Yellow Shape (Isolated)')
plt.axis('off')

plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 6))
plt.imshow(mask_green, cmap='gray')
plt.title('Green Shape (Isolated)')
plt.axis('off')
plt.show()
```

```python
# Task 04
import cv2
import matplotlib.pyplot as plt

# Load the shapes.bmp image
image = cv2.imread('./image_processing_files/shapes.bmp')

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Threshold the grayscale image to obtain binary masks for each shape
_, mask_red = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

# Find contours in the binary mask
contours, _ = cv2.findContours(mask_red, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Draw black circles at the center of each shape
for contour in contours:
    M = cv2.moments(contour)
    if M['m00'] > 0:
        cx = int(M['m10'] / M['m00'])
        cy = int(M['m01'] / M['m00'])
        cv2.circle(image, (cx, cy), 5, (0, 0, 0), -1)

# Plot the original image with center points marked
plt.figure(figsize=(6, 6))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Shapes with Center Points')
plt.axis('off')
plt.show()
```

```python
# Task 05
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the persp.jpg image
image = cv2.imread('./image_processing_files/persp.jpg')

# Define the four corners of the quadrilateral (hit and trial based)
src_points = np.float32([[230, 130], [470, 100], [50, 380], [650, 350]])

# Define the four outer corners where the quadrilateral will be mapped
dst_points = np.float32([[0, 0], [400, 0], [0, 300], [400, 300]])

# Compute the perspective transformation matrix
perspective_matrix = cv2.getPerspectiveTransform(src_points, dst_points)

# Apply the perspective transformation
result = cv2.warpPerspective(image, perspective_matrix, (400, 300))

# Display the original and transformed images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.title('Perspective Transformed Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the robot_green_bg image
robot_green_bg = cv2.imread('./image_processing_files/robot_green_bg.bmp')

# Check if the robot_green_bg image is loaded successfully
if robot_green_bg is None:
    print("Error: The robot_green_bg image was not loaded.")
else:
    # Define the green color range in HSV (you can adjust these values based on the specific shade of green)
    lower_green = (35, 50, 50)
    upper_green = (85, 255, 255)

    # Convert the image to HSV color space
    hsv = cv2.cvtColor(robot_green_bg, cv2.COLOR_BGR2HSV)

    # Create a mask to identify green pixels
    mask = cv2.inRange(hsv, lower_green, upper_green)

    # Replace green background with blue
    robot_green_bg[mask > 0] = [0, 0, 255]

    # Load the road image
    road = cv2.imread('./image_processing_files/road.jpg')

    # Check if the road image is loaded successfully
    if road is None:
        print("Error: The road image was not loaded.")
    else:
        # Resize the robot_green_bg image to match the dimensions specified for placement on the road image
        robot_green_bg_resized = cv2.resize(robot_green_bg, (300, 150))

        # Determine the position to place the robot on the road (you can adjust these values)
        x_offset, y_offset = 150, 100

        # Get the height and width of the resized robot image
        height, width = robot_green_bg_resized.shape[:2]

        # Place the robot on the road image
        road[y_offset:y_offset+height, x_offset:x_offset+width] = robot_green_bg_resized

        # Plot the modified robot_green_bg image and the robot on the road image
        plt.figure(figsize=(12, 6))

        plt.subplot(1, 2, 1)
        plt.imshow(cv2.cvtColor(robot_green_bg_resized, cv2.COLOR_BGR2RGB))
        plt.title('Modified Robot Green BG')
        plt.axis('off')

        plt.subplot(1, 2, 2)
        plt.imshow(cv2.cvtColor(road, cv2.COLOR_BGR2RGB))
        plt.title('Robot on Road')
        plt.axis('off')

        plt.tight_layout()
        plt.show()
```
